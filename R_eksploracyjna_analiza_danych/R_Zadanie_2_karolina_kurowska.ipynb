## PYTANIE 1 - Które kraje są największymi producentemi czekolady?

```r
# Tabela liczności dla kraju pochodzenia firmy
country_counts <- table(df2$Company_Location)

# Wykres słupkowy dla największych producentów czekolady z nazwami krajów
top_countries <- head(sort(country_counts, decreasing = TRUE), 10)
barplot(top_countries, main = "Najwięksi producenci czekolady",
        xlab = "Kraj", ylab = "Liczba firm", col = "skyblue",
        names.arg = names(top_countries), las = 2, cex.names = 0.7)
cat("Ilość firm w pierwszej dziesiatce krajów:")
print(top_countries)
```

W topce największych producentów czekolady na świecie są USA, Francja i Kanada.

## PYTANIE 3 - Jakie są trendy produkcji czekolady zwykłej i gorzkiej na rynku światowym? Czy któryś z tych rodzajów jest bardziej popularny?

```r
# Tabela liczności dla rodzaju czekolady
chocolate_type_counts <- table(df2$is_bitter)

# Wykres kołowy dla rodzaju czekolady
pie(chocolate_type_counts, main = "Popularność po rodzaju czekolady", labels = c("Zwykła", "Gorzka"), col = c("orange", "chocolate4"))

df <- read.csv("new_cocoa_karolina_kurowska.csv")

Is_Bitter_Yes <- df %>%
  filter(is_bitter == "Yes") %>%
  group_by(Review_Date) %>%
  summarise(Count = n())

Is_Bitter_No <- df %>%
  filter(is_bitter == "No") %>%
  group_by(Review_Date) %>%
  summarise(Count = n())


ggplot() +
  geom_line(data = Is_Bitter_Yes, aes(x = Review_Date, y = Count, color = "Gorzka")) +
  geom_line(data = Is_Bitter_No, aes(x = Review_Date, y = Count, color = "Zwykła")) +
  labs(title = "Popularność gorzkich i zwykłych czekolad",
       x = "Rok", y = "Ilość czekolad") +
  scale_color_manual(values = c("Gorzka" = "chocolate4", "Zwykła" = "orange")) +
  theme_minimal()
```

Na podstawie obu wykresów można wywnioskować, że:

Czekolada gorzka jest coraz bardziej popularna na świecie.
Czekolada zwykła jest coraz rzadziej kupowana

## PYTANIE 4 - Który kraj ma najwyższe oceny czekolady?

```r
# Średnia ocen dla każdego kraju
avg_ratings_by_country <- tapply(df2$Rating, df2$Company_Location, mean)

# Wybierz top 10 krajów z najwyższą średnią oceną
top_countries <- names(sort(avg_ratings_by_country, decreasing = TRUE)[1:10])

# Filtruj dane tylko dla top 10 krajów
df2_top <- df2[df2$Company_Location %in% top_countries,]

library(ggplot2)

# Stwórz wykres pudełkowy dla top 10 krajów według średniej oceny czekolady
ggplot(df2_top, aes(x = reorder(Company_Location, -Rating, FUN = median), y = Rating, fill = Company_Location)) +
  geom_boxplot() +
  labs(title = "Oceny czekolady według kraju",
       x = "Kraj",
       y = "Ocena") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust = 1),
        legend.position = "none")
```

Kraj z najwyższą średnią oceną: Chile
Średnia ocena: 3.75
Jednak biorąc pod uwagę kraje z większą ilością głosów lepszą średnią uzyskały australia i szwajcaria

## PYTANIE 6 – Jak zmieniały się oceny różnych rodzajów czekoladek na przestrzeni czasu?

```r
# Obliczenie średniej oceny dla każdego roku i typu
ratings_over_time <- df2 %>%
  group_by(Review_Date, is_bitter) %>%
  summarise(Average_Rating = mean(Rating, na.rm = TRUE), .groups = 'drop')

# Tworzenie wykresu
ggplot(ratings_over_time, aes(x = Review_Date, y = Average_Rating, color = is_bitter, group = is_bitter)) +
  geom_line() +
  geom_point() +
  scale_color_manual(values = c("No" = "orange", "Yes" = "chocolate4")) +
  labs(title = "Zmiany ocen różnych typów czekoladek w czasie",
       x = "Rok recenzji",
       y = "Średnia ocena",
       color = "Gorzka?") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

Wykres pokazuje, że czekolada gorzka miała wielki spadek ocen do roku  2007. Jednak od roku 2012 jej śrdnia ocena jest wyższa niż zwykłej. Wykes czekolady słodkiej jest mocno zmienny. Można jednak zauważyć, że do lat 2012, oceniana była średnio lepiej niż gorzka.
